#Langkah 2: Import Library & Inisialisasi Model

Di sel kode kedua, kita akan memanggil pipeline. Pipeline adalah fitur tingkat tinggi dari Hugging Face yang secara otomatis menangani:

Tokenisasi: Mengubah teks menjadi angka.

Arsitektur Model: Memuat struktur Transformer.

Prediksi: Menghasilkan output.

```python
# Langkah 2: Import pipeline
from transformers import pipeline
import pandas as pd # Untuk menampilkan hasil dalam tabel agar rapi

# Inisialisasi model analisis sentimen
# Secara default menggunakan model: distilbert-base-uncased-finetuned-sst-2-english
model_transformer = pipeline("sentiment-analysis")
print("Model Transformer berhasil dimuat!")
```

#Langkah 3: Membuat Fungsi Analisis & Data Uji

Kita akan membuat daftar kalimat percobaan dan sebuah fungsi untuk memprosesnya.

```python
# Langkah 3: Menyiapkan data teks untuk percobaan
data_uji = [
    "I am very happy with the results of this Transformer experiment!",
    "The documentation for this library is very confusing and hard to follow.",
    "The weather today is just okay, nothing special.",
    "This AI model is absolutely revolutionary and mind-blowing!"
]

def jalankan_prediksi(list_teks):
    hasil_raw = model_transformer(list_teks)

    # Gabungkan teks asli dengan hasil prediksi untuk visualisasi
    hasil_final = []
    for teks, res in zip(list_teks, hasil_raw):
        hasil_final.append({
            "Teks Masukan": teks,
            "Label": res['label'],
            "Skor Keyakinan": f"{res['score'] * 100:.2f}%"
        })
    return pd.DataFrame(hasil_final)

# Jalankan percobaan
df_hasil = jalankan_prediksi(data_uji)
```

#Langkah 4: Menampilkan Hasil Percobaan

Gunakan perintah ini untuk melihat hasil klasifikasi yang dilakukan oleh arsitektur Transformer.

```python
# Langkah 4: Tampilkan hasil dalam bentuk tabel yang estetik
print("--- HASIL PERCOBAAN TRANSFORMER ---")
df_hasil
```

#Langkah 5: (Interaktif) Input Teks Sendiri

Agar tugas Anda "Tuntas", kita tambahkan fitur di mana pengguna bisa memasukkan kalimat bebas untuk diuji oleh AI.

```python
# Langkah 5: Fitur Interaktif
print("\n=== Uji Coba Mandiri ===")
input_user = input("Masukkan kalimat dalam bahasa Inggris: ")
prediksi_user = model_transformer(input_user)

label = prediksi_user[0]['label']
skor = prediksi_user[0]['score'] * 100

print(f"\nAnalisis AI:")
print(f"Sentimen: {label}")
print(f"Keyakinan: {skor:.2f}%")
```

Penjelasan Teknis untuk Laporan Tugas Anda:
Jika Anda perlu menjelaskan apa yang terjadi di balik layar program Colab ini:

Tokenization: Kalimat Anda dipecah menjadi unit-unit kecil (token).

Self-Attention: Model melihat hubungan antar kata (Misal: kata "Revolutionary" sangat mempengaruhi sentimen positif).

Position Encoding: Model memahami urutan kata dalam kalimat tanpa menggunakan RNN.

Softmax Layer: Tahap akhir yang mengubah skor angka menjadi probabilitas (misal: 99% Positif).
